PHASE 1 : Model that recommends outfits based on occasion

```python
# Initialize Sentence-BERT model
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
# Encode product descriptions
data["text_data"] = data["productDisplayName"] + " " + data["json_data"]  # Merge relevant text
text_embeddings = model.encode(data["text_data"], show_progress_bar=True)

# Save embeddings for reuse
np.save("text_embeddings.npy", text_embeddings)
data.to_csv("fashion_data_with_embeddings.csv", index=False)
```

```python
# Load precomputed embeddings
text_embeddings = np.load("text_embeddings.npy")
```

```python
# Load image dataset
image_df = pd.read_csv("/content/images.csv")
image_df["filename"] = image_df["filename"].astype(str)
```

```python
# Ensure both columns are strings
data["id"] = data["id"].astype(str)
image_df["filename"] = image_df["filename"].astype(str).str.replace(".jpg", "", regex=False)

# Merge datasets
df = data.merge(image_df, left_on="id", right_on="filename", how="left")

# Drop redundant filename column
df.drop(columns=["filename"], inplace=True)

# Print first 5 rows to check
print(df.head())
```

```python
def recommend_outfits(user_input, top_n=5, min_similarity=0.6, season=None, base_color=None, gender=None, master_category=None, sub_category=None):
    """Finds top fashion items matching user input with optional filtering, ensuring results are always returned."""
    user_embedding = model.encode([user_input])  # Encode user query
    similarities = cosine_similarity(user_embedding, text_embeddings)[0]  # Compute similarity

    # Get top matches sorted by similarity
    top_indices = np.argsort(similarities)[::-1][:top_n]

    # Get recommended items
    recommendations = data.iloc[top_indices][["productDisplayName", "gender", "masterCategory",
                                              "subCategory", "articleType", "baseColour", "season"]].copy()

    # Apply optional filters
    if season:
        recommendations = recommendations[recommendations["season"].str.lower() == season.lower()]
    if base_color:
        recommendations = recommendations[recommendations["baseColour"].str.lower() == base_color.lower()]
    if gender:
        recommendations = recommendations[recommendations["gender"].str.lower() == gender.lower()]
    if master_category:
        recommendations = recommendations[recommendations["masterCategory"].str.lower() == master_category.lower()]
    if sub_category:
        recommendations = recommendations[recommendations["subCategory"].str.lower() == sub_category.lower()]

    # ✅ If all filters remove results, return top closest matches without filtering
    if recommendations.empty:
        return data.iloc[top_indices][["productDisplayName", "gender", "masterCategory",
                                       "subCategory", "articleType", "baseColour", "season"]]

    return recommendations
```

PHASE 2:Model that recommends pieces to complete the existing outfit

```python
# Load a general YOLOv8 model
model = YOLO("yolov8m.pt")
```

```python
image_paths = ["/content/outfit1.jpg", "/content/outfit2.jpg"]
```

```python
# Run detection on both images
results = [model(cv2.imread(img)) for img in image_paths]
```

```python
def get_dominant_colors(image, k=3):
    """Extracts the top k dominant colors from the image."""
    image = cv2.resize(image, (100, 100))  # Resize for faster processing
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB
    pixels = image.reshape(-1, 3)  # Reshape to 1D array

    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pixels)

    colors = kmeans.cluster_centers_.astype(int)  # Get dominant colors
    return colors
```

```python
def analyze_outfit(image_path):
    """Runs YOLO detection, and if no reliable results, extracts colors & patterns."""
    image = cv2.imread(image_path)
    results = model(image)  # Run YOLO detection

    detected_items = []
    for box in results[0].boxes:
        confidence = box.conf[0].item()
        if confidence >= 0.5:
            class_id = int(box.cls[0])
            detected_items.append(model.names[class_id])

    if detected_items:
        print(f"\n🔹 **Detected Clothing Items:** {detected_items}")
    else:
        print("\n⚠️ No reliable detections found! Extracting colors & patterns instead...")

        # Extract dominant colors
        colors = get_dominant_colors(image)
        print(f"🎨 **Dominant Colors:** {colors}")

        # Detect edges to check for patterns
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        edges = cv2.Canny(gray, 100, 200)  # Edge detection
        pattern_intensity = np.sum(edges) / edges.size  # Measure edge density

        pattern_type = "Solid" if pattern_intensity < 50 else "Patterned"
        print(f"🖼️ **Pattern Type:** {pattern_type}")
```

```python
# Test with images
image_paths = ["outfit1.jpg", "outfit2.jpg"]
for img_path in image_paths:
    analyze_outfit(img_path)
```

```python
def suggest_outfit_completions(detected_items, colors, pattern_type, occasion="Casual"):
    """Suggests missing outfit pieces based on detected clothing items."""

    detected_categories = set(data[data['articleType'].isin(detected_items)]['subCategory'])
    essential_categories = {"Topwear", "Bottomwear", "Footwear", "Accessories"}
    missing_categories = essential_categories - detected_categories

    suggestions = []
    for category in missing_categories:
        filtered_data = data[data['subCategory'] == category].copy()  # Copy to avoid SettingWithCopyWarning

        # Filter by occasion (if available)
        if "occasion" in data.columns:
            filtered_data = filtered_data[filtered_data["occasion"].str.contains(occasion, na=False, case=False)]

        # Sort by best color/pattern match
        if "baseColour" in data.columns:
            filtered_data.loc[:, "color_match"] = filtered_data["baseColour"].apply(
                lambda x: any(str(c) in str(x).lower() for c in colors)
            )
            filtered_data = filtered_data.sort_values(by="color_match", ascending=False)

        # Select top suggestion
        if not filtered_data.empty:
            suggestions.append(filtered_data.iloc[0]["productDisplayName"])

    return suggestions
```

```python
def analyze_outfit_and_suggest(image_path, occasion="Casual"):
    """Runs outfit analysis and suggests missing pieces."""
    image = cv2.imread(image_path)
    results = model(image)  # Run YOLO detection

    detected_items = []
    for box in results[0].boxes:
        confidence = box.conf[0].item()
        if confidence >= 0.5:
            class_id = int(box.cls[0])
            detected_items.append(model.names[class_id])

    if detected_items:
        print(f"\n🔹 **Detected Clothing Items:** {detected_items}")
    else:
        print("\n⚠️ No reliable detections found! Extracting colors & patterns instead...")
        detected_items = []

    # Extract dominant colors
    colors = get_dominant_colors(image)
    print(f"🎨 **Dominant Colors:** {colors}")

    # Detect patterns
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 100, 200)
    pattern_intensity = np.sum(edges) / edges.size
    pattern_type = "Solid" if pattern_intensity < 50 else "Patterned"
    print(f"🖼️ **Pattern Type:** {pattern_type}")

    # Suggest outfit completions
    suggestions = suggest_outfit_completions(detected_items, colors, pattern_type, occasion)
    print(f"\n✨ **Suggested Missing Pieces for {occasion}:** {suggestions}")
```

```python
# Test
analyze_outfit_and_suggest("outfit1.jpg", "Winter Party")
analyze_outfit_and_suggest("outfit2.jpg", "Casual Summer")
```
