```python
#import statements
import pandas as pd
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
import cv2
import json
```

```python
images_csv_path = "E:/Personal Projects/Fashion Recommendation System/archive/fashion-dataset/images.csv"
images_folder = "E:/Personal Projects/Fashion Recommendation System/archive/fashion-dataset/images"
```

```python
images_df = pd.read_csv(images_csv_path)
images_df.head()
```

```python
print(images_df.columns)
```

```python
missing_files = []
for index, row in images_df.iterrows():
    image_path = os.path.join(images_folder, row['filename'])
    image_path = image_path.replace("\\", "/")  # Normalize path separator
    if not os.path.exists(image_path):  # Check if the file exists
        missing_files.append(image_path)

# Print results
if missing_files:
    print(f"Found {len(missing_files)} missing files:")
    for file in missing_files[:10]:  # Show first 10 missing files
        print(file)
else:
    print("All image paths are valid!")
```

```python
valid_images_df = images_df[images_df['filename'].apply(lambda x: os.path.exists(os.path.join(images_folder, x.strip())))] 
```

```python
valid_images_df.to_csv('cleaned_images.csv', index=False)
```

```python
print(f"Original dataset size: {images_df.shape[0]} rows")
print(f"Cleaned dataset size: {valid_images_df.shape[0]} rows")
```

```python
output_folder = os.path.join(os.path.dirname(images_folder), 'output_images')
os.makedirs(output_folder, exist_ok=True)
print(f"'output_images' folder created at: {output_folder}")
```

```python
for index, row in tqdm(images_df.iterrows(), total=len(images_df)):
    input_path = os.path.join(images_folder, row['filename'])
    output_path = os.path.join(output_folder, row['filename'])
    if os.path.exists(input_path):
        image = cv2.imread(input_path)
        image_resized = cv2.resize(image, (224, 224))
        image_normalized = image_resized / 255.0
        image_to_save = (image_normalized * 255).astype('uint8')
        cv2.imwrite(output_path, image_to_save)
    else:
        print(f"Image not found: {input_path}")
print(f"Preprocessing complete. Preprocessed images saved in {output_folder}.")
```

```python
num_output_images = len([f for f in os.listdir(output_folder) if f.endswith('.jpg')])
print(f"Number of preprocessed images: {num_output_images}")
```

```python
styles_csv_path = "E:/Personal Projects/Fashion Recommendation System/archive/fashion-dataset/styles.csv"
styles_folder = "E:/Personal Projects/Fashion Recommendation System/archive/fashion-dataset/styles"
```

```python
styles_df = pd.read_csv(styles_csv_path, on_bad_lines='skip')
styles_df.head()
```

```python
missing_values = styles_df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
styles_df['baseColour'] = styles_df['baseColour'].fillna('Unknown')
styles_df['season'] = styles_df['season'].fillna('Unknown')
styles_df['productDisplayName'] = styles_df['productDisplayName'].fillna('Unknown')

styles_df['year'] = styles_df['year'].fillna(styles_df['year'].mean())

styles_df['usage'] = styles_df['usage'].fillna(0)

missing_values_after = styles_df.isnull().sum()
print("Missing values after filling:")
print(missing_values_after)
```

```python
print("Cleaned styles DataFrame:")
print(styles_df.head())
```

```python
styles_df.to_csv(styles_csv_path, index=False)
```

```python
def load_json_for_image(image_id):
    json_file = os.path.join(styles_folder, f'{image_id}.json')  # Assuming JSON filenames are based on the image id
    if os.path.exists(json_file):
        try:
            with open(json_file, 'r', encoding='utf-8') as f:
                return json.load(f)
        except UnicodeDecodeError as e:
            print(f"Error decoding {json_file}: {e}")
            return None  # or handle differently, depending on your use case
    else:
        return None  # or a default value 
```

```python
styles_df['json_data'] = styles_df['id'].apply(load_json_for_image)
```

```python
print(styles_df[['id', 'json_data']].head())
```

```python
def extract_json_features(json_str):
    try:
        data = json.loads(json_str)
        # Example: Extracting some metadata like product name and color from JSON
        features = {
            'product_name': data.get('meta', {}).get('product_name', None),
            'product_color': data.get('meta', {}).get('base_color', None),
            # Add other relevant fields here as per your requirements
        }
        return features
    except Exception as e:
        print(f"Error extracting data: {e}")
        return {}
```

```python
json_features = styles_df['json_data'].apply(extract_json_features)
```

```python
extracted_json_df = pd.json_normalize(json_features)
```

```python
styles_df = pd.concat([styles_df, extracted_json_df], axis=1)
```

```python
# Convert 'id' column to string in styles_df
styles_df['id'] = styles_df['id'].astype(str)

# Merge the DataFrames using 'id' from styles_df and 'filename' from images_df
styles_df = pd.merge(styles_df, images_df, left_on='id', right_on='filename', how='left')

# Save the merged DataFrame to a new CSV
styles_df.to_csv('merged.csv', index=False)

print("Merged CSV saved as 'merged.csv'")
```
